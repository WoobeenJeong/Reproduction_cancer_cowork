```python
import os
import h5py

# .h5 파일 열기
patches_dir = '/data/CLAM/LUAD_RESULT/patches/'
file_name = "TCGA-86-8076-01Z-00-DX1.e7378b2f-e20e-4d2f-a86c-3a8ead08a385.h5"
file_path = os.path.join(patches_dir, file_name)

with h5py.File(file_path, "r") as file:
    # 파일 내 그룹 및 데이터셋 확인
    for name in file:
        print(name)

```

```python
def print_hdf5_item(name, obj):
    """
    재귀적으로 HDF5 아이템을 탐색하고 출력하는 함수
    """
    if isinstance(obj, h5py.Group):  # 그룹인 경우
        print(name + "/ (Group)")
        for key in obj.keys():
            print_hdf5_item(name + "/" + key, obj[key])
    elif isinstance(obj, h5py.Dataset):  # 데이터셋인 경우
        print(name + " (Dataset):")
        print(obj[()])  # 데이터셋의 내용 출력

with h5py.File(file_path, "r") as file:
    file.visititems(print_hdf5_item)
```
